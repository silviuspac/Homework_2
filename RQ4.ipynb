```python
import pandas as pd
import ujson
import codecs
```

```python
events = pd.read_json('events_England.json')
df_events = pd.DataFrame(events, columns = ['eventId', 'tags', 'playerId', 'eventName'])
df_events = df_events.loc[df_events['eventId'] == 8]
df_events = df_events.assign(comp = df_events["tags"].apply(won_duel))
df_events
```

```python
def won_duel(tags):
    for el in tags:
        if el['id'] == 1801:
            return 1
    return 0
```

```python
players = pd.read_json('players.json')
df_players = pd.DataFrame(players, columns = ['wyId', 'shortName'])
df_players
```

```python
df_players.rename(columns = {'wyId' : 'playerId'}, inplace = True)
df_merged = pd.merge(df_events, df_players, on = "playerId")
df_merged
```

```python
df_merged.groupby(['shortName']).size().loc[lambda group: group > 0]
```

```python
#df_supp = df_merged.groupby(['shortName', 'comp']).size().loc[lambda group: group > 1200]
value_counts = df_merged['shortName'].value_counts()
to_remove = value_counts[value_counts < 1100].index
df_test = df_merged[~df_merged.shortName.isin(to_remove)]
df_test
```

```python
df_test['shortName'].value_counts()
```

```python
df_g = df_test[df_test['comp'] == 1].groupby(['shortName']).size()/(df_test.groupby(['shortName']).size())
df_g
```

```python
df_g.sort_values(ascending = False, inplace = True)
res = df_g.to_dict()
l = list(res.keys())
for i in range(10):
    l[i] = [codecs.unicode_escape_decode(l[i])[0], df_g[i]]
    print(f"{l[i][0]:<15}{l[i][1]:>10}")
          
#           .format(l[i][0], l[i][1]))
    
    
# Sono mostrati solo i giocatori con presenze totali maggiori della media di presenze Premier League IDEONA
```

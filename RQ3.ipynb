```python
import pandas as pd
```

```python
coaches = pd.read_json("../data/coaches.json")
teams = pd.read_json("../data/teams.json")
df_coaches = pd.DataFrame(coaches, columns = ['wyId','currentTeamId','birthDate'])
df_teams = pd.DataFrame(teams, columns = ['wyId','officialName','area'])
for i in range(len(df_teams)):
    df_teams.area[i] = df_teams.area[i]['name']
df_teams = df_teams.loc[df_teams['area'] == 'England']
#df_coaches
df_teams
```

```python
df_teams.rename(columns = {'wyId' : 'TeamId'}, inplace = True)
df_coaches.rename(columns = {'currentTeamId' : 'TeamId', 'wyId' : 'CoachId'}, inplace = True)
df_merged = pd.merge(df_teams, df_coaches, on = "TeamId")
df_merged
```

```python
df_merged.sort_values(by = ['birthDate'], ascending = False, inplace = True)
df_merged.reset_index(inplace = True)
df_merged
```

```python
df_merged.drop_duplicates(subset='officialName', keep='first', inplace = True)
df_merged
```
